# DATA VISUALIZATION

```python
#checking correlation between columns through Heatmaps using seaborn library
f,ax=plt.subplots(figsize=(12,8))
sns.heatmap(corr,annot=True,cmap='Blues',linewidths=0.9)
plt.show()
```

```python
#Visualization using histogram
analysis_data.hist(bins=40,figsize=(6,8))
plt.show()
```

**FROM THE ABOVE HISTOGRAM WE INFER THAT OUR DATASET IS NOT NORMAL**

```python
# segment wise count of ship modes using seaborn
sns.catplot(x='Ship Mode',hue='Segment',data=analysis_data,kind='count',height=7,aspect=2.5,palette='Set2')
plt.show()
```

**AS WE SEE CONSUMER SEGMENT USES MOST STANDARD CLASS AS THE MODE OF SHIPMENT AND THIS SEGMENT IS HIGHER IN EVERY MODE OF SHIPMENT**

```python
#Segment wise sales in each region
plt.figure(figsize=(10,8))
sns.barplot(x='Region',y='Sales',hue='Segment',data=analysis_data)
plt.show()
```

**AS WE SEE NOT MUCH DIFFERENCE IN SALES PER REGION OF SEGMENT.A LITTLE EXTRA CORPORATE IS HIGHER PER CENTRAL REGION SALES  AS COMPARED WITH OTHER TWO.**

```python
#ITEMS THAT SALE MORE IN SUBCATEGORY
analysis_data['Sub-Category'].value_counts().plot.pie(figsize=(10,8),autopct='%.2f',shadow=True)
plt.show()

```

**FROM THE ABOVE PIE CHART WE OBSERVED THAT BINDERS HAS THE MOST SALES IN SUBCATEGORY FOLLOWED BY PAPER,FURNITURE AND PHONES ETC**

**GETTING INFO ABOUT CATEGORY HAVING HIGHER SALES REGION WISE**

```python
pd.crosstab(index=analysis_data.Region,columns=analysis_data.Category).plot(kind='bar',figsize=(10,8))
```

## SO WE GET AN IDEA FROM THE ABOVE BAR CHART THAT CATEGORY OFFICE SUPPLIES ARE LEADING IN EVERY REGION,WEST US HAS MAXIMUM SALES AND SOUTH US HAVE MINIMUM SALES

**NOW USING STATS AND REPORTS WE ANALYZE TOTAL SALES CITY WISE AND PROFIT AND LOSS CITY WISE ALSO**

```python
grouped=analysis_data.groupby('City')
grouped.Sales.sum().sort_values(ascending=False).head().reset_index()
```

#### AS WE SEE FROM ABOVE REPORT THAT NEW YORK CITY IS LEADING IN SALES FOLLOWED BY LOS ANGELES AND LEAST SALES IS DONE BY PHILADELPHIA CITY.

```python
Less_profit=grouped.Profit.sum().sort_values(ascending=True).head().reset_index()
```

```python
More_profit=grouped.Profit.sum().sort_values(ascending=False).head().reset_index()
```

```python
print('Less profit cities\n',Less_profit)
print('------'*6)

print('More profit cities\n',More_profit)
```

```python
#PLOTTING BAR GRAPHS OF PROFITABLE AND LOSS MAKING CITIES OF USA
fig,axes=plt.subplots(1,2,figsize=(10,6))
grouped.Profit.sum().sort_values(ascending=True).head().plot(kind='bar',ax=axes[0])
axes[0].tick_params(labelrotation=20)
axes[0].set_title('Loss making cities or Minimum Profit Cities')
plt.xticks(rotation=30)
grouped.Profit.sum().sort_values(ascending=False).head().plot(kind='bar',ax=axes[1])
axes[1].set_title('Profit Making Cities')
plt.show()
```

## SO AS WE INFER FROM ABOVE BAR GRAPH THAT 'NEW YORK CITY' MAKES HIGHEST PROFIT OUT OF ALL CITIES AND 'PHILADELPHIA' MAKES HIGH LOSS OR LOW PROFIT OUT OF ALL CITIES

### **NOW WE SEE STATE WISE SALES** 

```python
grpby_state=analysis_data.groupby('State').Profit.sum()
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x=grpby_state.index,y=grpby_state.values)
plt.xticks(rotation=90)
plt.title('State Wise Profit')
```

### HENCE WE SEE THAT CALIFORNIA STATE HAS MAX PROFIT ANDTEXAS HAS MAXIMUM LOSS

# PLOTTING REGION WISE PROFITS THROUGH PIE_CHART

```python
region_profit=analysis_data.groupby('Region').Profit.sum()
region_profit_sort = region_profit.sort_values(ascending = False).head(10)
```

```python
plt.pie(region_profit_sort, labels = region_profit_sort.index, autopct= "%.2f", explode = (0.05,0.05,0.05,0.20),radius=1.5)

centre_circle = plt.Circle((0,0), 0.90, fc = 'white')
fig= plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

### SO WE GET WEST REGION HAS HIGHEST PROFIT
